### 2.支持向量机训练

使用scikit-learn库中的支持向量机，分别使用线性核和高斯核进行训练

```python
# -*- coding: utf-8 -*-
#数据加载与预处理
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_table('D:\\Env\\vscode\\DeepLearning\\SVM\\watermelon.txt',delimiter=',')    
x = pd.DataFrame({'密度':data['密度'],'含糖率':data['含糖率'],'触感':data['触感'],
                    '脐部':data['脐部'],
                    '纹理':data['纹理'],
                    '敲声':data['敲声'],
                    '根蒂':data['根蒂'],
                    '色泽':data['色泽']})
x = x.values.tolist()    
encoder = LabelEncoder()
y = encoder.fit_transform(data['好瓜']).tolist()
x,y = np.array(x),np.array(y)
```

```python
#线性核
linear_svm = svm.SVC(C=0.5,kernel='linear')
linear_svm.fit(x,y)
svl = linear_svm.support_vectors_
print(svl,end='\n');
#支持向量如下：
```

```python
gauss_svm = svm.SVC(C=0.5,kernel='rbf')
gauss_svm.fit(x,y)
svg = gauss_svm.support_vectors_
print(svg,end='\n')
#支持向量如下:
```

可以看到高斯核训练出的支持向量机支持向量数量较线性核更多，可以推测高斯核支持向量机分类准确度较线性核更好。
